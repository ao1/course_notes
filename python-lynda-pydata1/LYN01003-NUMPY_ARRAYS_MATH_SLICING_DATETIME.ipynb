### math operations

```python
# make array from 0 to 10, 40 numbers
g = np.linspace(0,10,40)
```

```python
# make array that applies sin function to the array
sing = np.sin(g)
```

```python
# make array that applies cos function to the array
cosg = np.cos(g)
```

```python
# plot both arrays
plt.plot(g,sing)
plt.plot(g,cosg)
```

```python
# give different symbols to the two lines
plt.plot(g,sing,'x')
plt.plot(g,cosg,'o')
```

```python
y = sing + cosg
z = cosg**2 - sing**2

plt.plot(g,y)
plt.plot(g,z)
plt.legend(['sin(x) cos(x)','cos(x)^2 - sin(x)^2'])
```

```python
# take the inner product of 2 vectors
# (the sum of the element by element products)
np.dot(sing,cosg)
```

```python
# take the outer product of 2 vectors
# (builds every combination of the 
# elements from the two vectors)
np.outer(sing,cosg)
```

```python
# create a vector

h = np.linspace(0,10,5)
h
```

```python
# add 1 to the vector

h + 1
```

```python
# create a 2d array

hh = np.outer(h,h)
hh
```

```python
# adding 1 to the 2d array
hh + 1
```
