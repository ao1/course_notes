### DataFrames - merging

```python
tacos = pd.read_csv('tacos.csv')
tacos.head(2)
```

```python
tacos.shape # col , rows
```

```python
taco_toppings = pd.read_csv('taco_toppings.csv')
taco_toppings.head(2)
```

```python
taco_toppings.shape
```

```python
#both dataframes have a common column: name
#merge will automatically infer which column is the key
pd.merge(tacos,taco_toppings)
```

```python
#more common syntax:
tacos.merge(taco_toppings)
```

```python
tacos.merge(taco_toppings,how='outer')
```
